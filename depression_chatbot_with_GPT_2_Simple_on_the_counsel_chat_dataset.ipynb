```python
import pandas as pd
```

```python
Data = pd.read_csv("https://raw.githubusercontent.com/nbertagnolli/counsel-chat/master/data/counselchat-data.csv")
```

```python
text_data = ""
```

```python
file = open("data.txt", "w")
```

```python
for question, answer in zip(Data["questionText"].values, Data["answerText"].values):
  question = str(question)
  answer = str(answer)
  text_data += "<!question!> " + question + "<!/question!>" + "<!answer!>" + answer + "<!/answer!>"
```

```python
len(text_data)
```

```python
file.writelines(text_data)
```

```python
file.close()
```

```python
import gpt_2_simple as gpt
```

```python
model_name = "124M"
```

```python
gpt.download_gpt2(model_name=model_name)
```

```python
file_name = "data.txt"
```

```python
sess = gpt.start_tf_sess()
gpt.finetune(sess,
              file_name,
              model_name=model_name,
              steps=1000)   # steps is max number of training steps

gpt.generate(sess)
```

```python
gpt.generate(
    sess, prefix = "<!question!>i am hate my self, i feel like am making no progress fuuuck <!/question!><!answer!>"
)
```
